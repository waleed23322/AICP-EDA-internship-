```python
import matplotlib.pyplot as plt

# Define the x and y axis values
x_values = [1, 4, 7, 10]
y_values = [2, 5, 8, 11]

# Plot the line
plt.plot(x_values, y_values)

# Set the axis labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Set the title of the plot
plt.title("Line Plot")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for line 1
x1 = [10, 20, 30, 40, 50]
y1 = [30, 35, 25, 30, 20]

# Data for line 2
x2 = [10, 20, 30, 40, 50]
y2 = [20, 25, 35, 20, 25]

# Plot lines with different widths and colors
plt.plot(x1, y1, label='line1-width-3', linewidth=3, color='blue')
plt.plot(x2, y2, label='line2-width-5', linewidth=5, color='red')

# Add legend
plt.legend()

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two or more lines with different widths and colors with suitable legends')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for line 1
x1 = [2, 4, 6, 8, 10]
y1 = [3, 5, 7, 9, 11]

# Data for line 2
x2 = [2, 4, 6, 8, 10]
y2 = [3.5, 5.5, 7.5, 9.5, 11.5]

# Data for line 3
x3 = [2, 4, 6, 8, 10]
y3 = [4, 6, 8, 10, 12]

# Create the plot
plt.plot(x1, y1, label='Line 1', marker='o', markersize=5, color='b')  # blue circles
plt.plot(x2, y2, label='Line 2', marker='s', markersize=5, color='g')  # green squares
plt.plot(x3, y3, label='Line 3', marker='^', markersize=5, color='r')  # red triangles

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot with markers')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Programming languages
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']

# Popularity (replace with your data)
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Create the bar chart
plt.barh(languages, popularity, color='skyblue')

# Add labels and title
plt.xlabel('Popularity')
plt.ylabel('Programming Language')
plt.title('Popularity of Programming Languages\n(October 2017)')

# Customize the chart
plt.gca().invert_yaxis()  
plt.grid(axis='x', linestyle='--', alpha=0.6)  
plt.tight_layout()  

# Show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Country": ["China", "India", "USA", "Indonesia", "Brazil", "Nigeria", "Bangladesh", "Russia", "Japan", "Mexico"],
    "Employees": [1800, 1500, 1300, 1100, 800, 700, 650, 600, 550, 450]
}

# Load data into DataFrame
df = pd.DataFrame(data)

# Create the bar chart
plt.bar(df["Country"], df["Employees"], color='skyblue')

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Number of Employees')
plt.title('Number of Employees from Different Countries')

# Customize the chart
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Country names and their gold medal achievements
countries = ["United States", "Great Britain", "China", "Russia", "Germany"]
gold_medals = [46, 27, 26, 19, 17]

# Create the pie chart
plt.pie(gold_medals, labels=countries, autopct="%1.1f%%")

# Add a title
plt.title("Gold Medal Achievements of Top 5 Countries in 2016 Summer Olympics")

# Show the chart
plt.show()
```
