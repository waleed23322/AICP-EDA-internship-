```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Read the CSV file
df = pd.read_csv('E:\AICP EDA\Week4\EDA Internship 2.0 Week 4\menu.csv')
```

```python
# Check statistical facts
stats = df.describe()
print(stats)

# Calculate maximum values
max_values = df[['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 
                 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 
                 'Iron (% Daily Value)']].max()
print("Maximum values:")
print(max_values)
```

```python
# Plot correlation matrix
corr_matrix = df[['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 
                  'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 
                  'Iron (% Daily Value)']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Draw boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Calories', data=df)
plt.xticks(rotation=90)
plt.title('Calories vs Category')
plt.show()
```

```python
# Find item with maximum values
max_item = df.loc[df[['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 
                      'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 
                      'Iron (% Daily Value)']].idxmax()]
print("Item with maximum values:")
print(max_item)
```

```python
# Draw Stripplot for each category against each attribute
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 
              'Iron (% Daily Value)']

for attribute in attributes:
    sns.stripplot(x='Category', y=attribute, data=df, jitter=True)
    plt.title(f'{attribute} by Category')
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Draw a horizontal bar graph for items in each category against calories
categories = df['Category'].unique()

for category in categories:
    plt.figure(figsize=(10, 6))
    subset = df[df['Category'] == category]
    sns.barplot(x='Calories', y='Item', data=subset)
    plt.title(f'Calories in {category}')
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.show()
```
